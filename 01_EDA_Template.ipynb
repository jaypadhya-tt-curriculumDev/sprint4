## Load Test Data

Loading the Titanic dataset from a public URL to test data loading.

```
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(f'Dataset shape: {df.shape}')
print()
df.head()
```

```
# SQLite test
conn = sqlite3.connect(':memory:')
df.to_sql('passengers', conn, index=False)

result = pd.read_sql('SELECT Pclass, COUNT(*) as count FROM passengers GROUP BY Pclass', conn)
print(result)
```
